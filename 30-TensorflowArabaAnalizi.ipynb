```
y = dataFrame["price"].values

x = dataFrame.drop("price" ,axis=1).values
```

```
y
```

```
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 10) 
# random_state test ve traine bölümümüzün rastgele olacak şekilde yapıldığı ancak ++ 
# aynı zamanda tekrar aynı rastgele verileri almak istyiorsak kullancağımızı değişkende.
```

```
len(x_train)
```

```
len(x_test)
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
scaler = MinMaxScaler()
```

```
x_train = scaler.fit_transform(x_train)
```

```
x_test = scaler.transform(x_test)
```

```
from tensorflow.keras.models import Sequential # modelimizi oluşturuyoruz
from tensorflow.keras.layers import Dense # katmanlarımızı oluşturuyoruz
```

```
x_train.shape
```

```
model = Sequential()

model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))

model.add(Dense(1)) # Bu çıkış katmanı yukarıdakiler derin ağ katmanı

model.compile(optimizer = "adam", loss = "mse")
```

```
# model eğitilirken valide edilebiliyor bunun için validation_data ya x_test ve y_testi tuple şeklinde veriyoruz
# bach_size : verilerin hepsini birden modele vermek modelimizi yorucaktır bu yüzden ölerek veriyoruz örneğimizde 250 vereceğiz
# 250 yi oynatabiliriz modelin işleyişini bozmaz ancak epoch modelimizin gelen işleyişini bozabilir ancak bach_size sadece modelin öğrenmesi yavaşlar
model.fit(x = x_train, y = y_train, validation_data= (x_test, y_test), batch_size = 250 ,epochs = 300)
```

```
# val_loss validation_data den geliyor anlamı doğrulama kayıpları
# doğrulama kayıplarını da grafikleştirebileceğiz

kayipVerisi = pd.DataFrame(model.history.history)
```

```
kayipVerisi
```

```
kayipVerisi.plot()
```

```
# yukarıdaki grafik güzel ama belli bir yerden sonra loss veya val_loss' un ayrıldığını görseydik o zaman sıkıntı var demektir
# eğer böyle bir sıkıntı old. zaman overfeeding old. anlamına gelir ve epochs'u azalmamız gerekir veya başka işlemler yapmamız lazım
```

```
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```
tahminDizisi = model.predict(x_test)
```

```
tahminDizisi
```

```
mean_absolute_error(y_test, tahminDizisi)
```
